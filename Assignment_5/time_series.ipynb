```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Storing the data into 5 dataframes, the parse dates function is used to convert strings to a date datatype
stock_GOOG = pd.read_csv('AAPL.csv', parse_dates=['Date'])
stock_AAPL = pd.read_csv('AMZN.csv', parse_dates=['Date'])
stock_MSFT = pd.read_csv('GOOG.csv', parse_dates=['Date'])
stock_FCBK = pd.read_csv('META.csv', parse_dates=['Date'])
stock_AMZN = pd.read_csv('MSFT.csv', parse_dates=['Date'])
```

```python
# Keeping only the two columns for each dataframe
stock_GOOG = stock_GOOG[['Date', 'Close']]
stock_AAPL = stock_AAPL[['Date', 'Close']]
stock_MSFT = stock_MSFT[['Date', 'Close']]
stock_FCBK = stock_FCBK[['Date', 'Close']]
stock_AMZN = stock_AMZN[['Date', 'Close']]
```

```python
# Now we can use the first column as index 
stock_GOOG = stock_GOOG.set_index('Date')
stock_AAPL = stock_AAPL.set_index('Date')
stock_MSFT = stock_MSFT.set_index('Date')
stock_FCBK = stock_FCBK.set_index('Date')
stock_AMZN = stock_AMZN.set_index('Date')
```

```python
stock_AMZN.head()
```

```python
# Renaming the series
stock_GOOG = stock_GOOG.rename(columns = {'Close' : 'GOOGL'})
stock_AAPL = stock_AAPL.rename(columns = {'Close' : 'AAPL'})
stock_MSFT = stock_MSFT.rename(columns = {'Close' : 'MSFT'})
stock_FCBK = stock_FCBK.rename(columns = {'Close' : 'META'})
stock_AMZN = stock_AMZN.rename(columns = {'Close' : 'AMZN'})
```

```python
stock_MSFT.head()
```

```python
# Now we can combine the series into one dataframe
data_all = stock_GOOG.merge(stock_AAPL, left_index = True, right_index = True)
data_all = data_all.merge(stock_MSFT, left_index = True, right_index = True)
data_all = data_all.merge(stock_FCBK, left_index = True, right_index = True)
data_all = data_all.merge(stock_AMZN, left_index = True, right_index = True)
data_all = data_all.dropna()
data_all.head()
```

```python
# We can represent the time series in a graph 
plt.clf()

plt.plot(data_all, label = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN'])
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Evolution of the price over time")
plt.legend()
plt.show()
```

```python
# We use resample function - The "last" method takes the last value, any other aggregation could have been used (sum, mean...)
data_all_month = data_all.resample('M').last()
# M for month
data_all_month.head()
```

```python
# We can represent the time series in a graph 
plt.clf()
plt.plot(data_all_month, label = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN'])
plt.xlabel("Month")
plt.ylabel("Price")
plt.title("Evolution of the price over the months")
plt.legend()
plt.show()
```

### Usage of time series in Finance

```python
# Calculating the returns
data_returns = data_all.pct_change()
data_returns = data_returns.dropna()
data_returns.head()
# It is important to remember that the returns are percentages
```

```python
# Let us show the reason why returns can be used but are not the most suitable
plt.clf()
plt.plot(data_returns, label = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN'])
plt.xlabel("Month")
plt.ylabel("Return")
plt.title("Evolution of the returns")
plt.legend()
plt.show()
```

```python
# Calculating the cumulative returns 
data_cumulative_returns = (1 + data_returns).cumprod() - 1
data_cumulative_returns.head()
```

```python
# Displaying the cumulative returns
plt.clf()
plt.plot(data_cumulative_returns, label = ['GOOGL', 'AAPL', 'MSFT', 'META', 'AMZN'])
plt.xlabel("Month")
plt.ylabel("Cumulative returns")
plt.title("Evolution of the stock price in percentage over time")
plt.legend()
plt.show()
```

#### Select the data of the last 5 years
